#Plotting the scatter plot

Extracting features for plotting

```python
X=data_features['PetalWidthCm']
y=data_features['SepalWidthCm']
```

```python
X
```

```python
y
```

Plotting the features

```python
plt.scatter(x=X,y=y,color='blue')
plt.xlabel('Petal width(cm)')
plt.ylabel('Sepal width(cm)')
plt.title('Scatter plot of sepal width vs petal width')
plt.show()
```

#Plotting the line plot

Creating array of index of all datapoints

```python
datapoints_index=[]
for i in range(len(data)):
  datapoints_index.append(i)
```

Plotting the line plot for each feature

```python
fig, ax = plt.subplots()    # Creating axes

for feature in data_features.columns:                            # Iterating over each feature
    ax.plot(datapoints_index, data[feature], label=feature)      # Function to plot the datapoints

ax.set_title('Line plot for Iris Dataset')
ax.legend()
```

# Plotting the pair plot

```python
sns.pairplot(data) #library function provided by sns to plot pairplot
```

#Plotting 3D plot

```python
fig = px.scatter_3d(data, x='PetalWidthCm', y='PetalLengthCm', z='SepalLengthCm', color='Species')      # Library function to plot 3d scatter plot
fig.show()
```

#Plotting Histogram

```python
fig,ax=plt.subplots() #creating axes and figures
ax.hist(data_features['PetalWidthCm'])

ax.set_title('Histogram of Petal Width(cm)')
ax.set_xlabel('Petal width(cm)')
ax.set_ylabel('Frequency of datapoints')
```
